```python
import plotly.express as px
import pandas as pd
import mysql.connector;
        

def connect():
    user=input("user: ")
    password=input("\npassword: ")
    
    try: #connexion
        mydb = mysql.connector.connect(
                                      host="localhost",
                                      user=str(user),
                                      passwd=str(password),
                                      #database="Vehicule_CO2_Emmission",
                                      auth_plugin="mysql_native_password"
                                    )
        return mydb
        
    except:
        print("échec de connexion")
        return None
    

mydb=connect()
```

```python
def drop():
    mycursor=mydb.cursor()
    
    command='''DROP SCHEMA IF EXISTS Vehicule_CO2_Emmission;'''
    mycursor.execute(command)
    mydb.commit()
    #------------------------------------------------------------
    command='''CREATE DATABASE Vehicule_CO2_Emmission;'''
    mycursor.execute(command)
    mydb.commit()

    #------------------------------------------------------------
    command='''USE Vehicule_CO2_Emmission; '''
    mycursor.execute(command)
    mydb.commit()
    
def create_database(sql_file):
    mycursor=mydb.cursor()
    #------------------------------------------------------------
    with open(sql_file, 'r') as sql_file:
        result_iterator = mycursor.execute(sql_file.read(), multi=True)
        #for res in result_iterator:
        #    print("Running query: ", res)  # Will print out a short representation of the query
        #    print(f"Affected {res.rowcount} rows" )
        mydb.commit()  # Remember to commit all your changes!
        #sql_file.close()
    print("remplissage réussi")
```

```python
#----------------------
# data table Marque:
#----------------------
Marque=[]
Modele_com=[]

#----------------------
# data table Voiture:
#----------------------
Designation_com=[]
Cnit=[]

#----------------------
# data table Energie:
#----------------------
Carburant=[]
Hybride=[]

#----------------------
# data table Puissance:
#----------------------
Puissance_admin=[]
Puissance_max=[]
Boite_vitesse=[]
Masse=[]

#----------------------
# data table Consommation:
#----------------------
Conso_urb=[]
Conso_exurb=[]
Conso_mixte=[]

#----------------------
# data table emmissions:
#----------------------
Co2=[]
CO =[]
HC =[]
NOx=[]
HC_NOX=[]
Particules=[]

#----------------------
# data table Carrosserie:
#----------------------
Carrosserie=[]

#----------------------
# data table Gamme:
#----------------------
Gamme=[]


fichier="mars-2014-complete.csv"

f = open(fichier, "r+")     # mode lecture/écriture
lignes = f.readlines()
f.seek(0)                         # retour au début du fichier
for i in range(len(lignes)):
    if i!=0:
        # découper les lignes en plusieurs éléments
        line = lignes[i].split(";")
        
        #----------------------
        # data table Marque:
        #----------------------
        Marque.append(line[0])
        Modele_com.append(line[1])
        
        #----------------------
        # data table Voiture:
        #----------------------
        Designation_com.append(line[3])
        Cnit.append(line[4])
        
        #----------------------
        # data table Energie:
        #----------------------
        Carburant.append(line[6])
        Hybride.append(line[7])
        
        #----------------------
        # data table Puissance:
        #----------------------
        Puissance_admin.append(line[8])
        Puissance_max.append(line[9])
        Boite_vitesse.append(line[10])
        Masse.append(line[20])
        
        #----------------------
        # data table Consommation:
        #----------------------
        Conso_urb.append(line[11])
        Conso_exurb.append(line[12])
        Conso_mixte.append(line[13])
        
        #----------------------
        # data table emmissions:
        #----------------------
        Co2.append(line[14])
        CO.append(line[15])
        HC.append(line[16])
        NOx.append(line[17])
        HC_NOX.append(line[18])
        Particules.append(line[19])
        
        
        #----------------------
        # data table Carrosserie:
        #----------------------
        Carrosserie.append(line[24])

        #----------------------
        # data table Gamme:
        #----------------------
        Gamme.append(line[25])
        
f.close()
```

```python
import pandas as pd

data_dict={"Cnit":Cnit,"Marque":Marque,"Modele_com":Modele_com,"Designation_com":Designation_com,"Carburant":Carburant,
           "Hybride":Hybride,"Puissance_admin":Puissance_admin,"Puissance_max":Puissance_max,"Boite_vitesse":Boite_vitesse,
           "Masse":Masse,"Conso_urb":Conso_urb,"Conso_mixte":Conso_mixte,"Conso_exurb":Conso_exurb,"Co2":Co2,"CO":CO,
           "HC":HC,"NOx":NOx,"HC_NOX":HC_NOX,"Particules":Particules,"Carrosserie":Carrosserie,"Gamme":Gamme}
data=pd.DataFrame.from_dict(data_dict)
```

```python
data=data.drop_duplicates(subset=['Cnit'])
data=data.set_index([pd.Index([i for i in range(data.shape[0])])])
print(data.shape)
```

```python
def clearTable(table_name: str):
    mycursor=mydb.cursor()
    command=f'''delete from {table_name}'''
    print(command)
    mycursor.execute(command)
    mydb.commit()
    

def addDataToTable(table_name: str, data: pd.DataFrame):
    mycursor = mydb.cursor()
    data=data.replace("",0)
    data=data.set_index([pd.Index([i for i in range(data.shape[0])])])
    
    #------------------------------
    # table marque:
    #------------------------------
    
    if table_name=="Marque":
        Marque_Text = open(r'./DATA_BASE_CSV/Marque_data.txt','w')
        command= ""
        #try:
        for i in data.index.to_list():
            marque=data.loc[i,"Marque"]
            Modele_com=data.loc[i,"Modele_com"]
            Designation_com=data.loc[i,"Designation_com"]

            command+=f'''insert into Marque(ID_Marque,Marque,Modele_com, Designation_com) Values({i},"{marque}","{Modele_com}","{Designation_com}");\n'''

            #mycursor.execute(command)
            #mydb.commit()

        Marque_Text.write(command)
        Marque_Text.close()
        #mycursor.close() 
        #except:
        #print("error table Marque ou  déjà complétée")
            
            
    #------------------------------
    # table Gamme:
    #------------------------------
    elif table_name=="Gamme":
        data=data.set_index([pd.Index([i for i in range(data.shape[0])])])
        command=""
        #try:
        for i in data.index.to_list():
            gamme=data.loc[i,"Gamme"].replace('\n','')

            command+=f'''insert into Gamme(ID_Gamme,Gamme) Values({i},"{gamme}");'''


        mycursor.execute(command)
        mydb.commit()  
        #except:
        #    print("error table Gamme ou  déjà complétée")
            
    #------------------------------
    # table Energie:
    #------------------------------
    elif table_name=="Energie":
        hybride_dict={"non":0,"oui":1}
        data=data.set_index([pd.Index([i for i in range(data.shape[0])])])
        #try:
        for i in data.index.to_list():
            Carburant=data.loc[i,"Carburant"]
            Hybride=hybride_dict[data.loc[i,"Hybride"]]

            command=f'''insert into Energie(ID_Energie,Carburant, Hybride) Values({i},"{Carburant}","{Hybride}")'''
           
            mycursor.execute(command)
            mydb.commit()
            
        mycursor.close() 
        #except:
        #print("error table Energie ou  déjà complétée")
        
        
    #------------------------------------------------------------
    # Table: Consommation
    #------------------------------------------------------------
    elif table_name=="Consommation":
        Consommation_Text = open(r'./DATA_BASE_CSV/Consommation_data.txt','w')
        command= ""
        #try:
        for i in data.index.to_list():
            Conso_urb=data.loc[i,"Conso_urb"]
            Conso_exurb=data.loc[i,"Conso_exurb"]
            Conso_mixte=data.loc[i,"Conso_mixte"]


            command+=f'''insert into `Consommation` (`ID_Consommation`,`Conso_urb`, `Conso_exurb`, `Conso_mixte`) Values({i},{Conso_urb},{Conso_exurb},{Conso_mixte});\n'''

            #mycursor.execute(command)
            #mydb.commit()
        Consommation_Text.write(command)
        Consommation_Text.close() 
        #mycursor.close() 
        #except:
        #print("error table Consommation ou  déjà complétée")
    
    #------------------------------
    # table Puissance:
    #------------------------------
    elif table_name=="Puissance":
        data=data.set_index([pd.Index([i for i in range(data.shape[0])])])
        try:
            for i in data.index.to_list():
                Puissance_max=data.loc[i,"Puissance_max"]
                Puissance_admin=data.loc[i,"Puissance_admin"]
                Boite_vitesse=data.loc[i,"Boite_vitesse"]
                Masse=data.loc[i,"Masse"]

                command=f'''insert into Puissance(ID_Puissance,Puissance_max, Puissance_admin, Boite_vitesse, Masse) Values({i},{Puissance_max},{Puissance_admin},'{Boite_vitesse}',{Masse})'''         

                mycursor.execute(command)
                mydb.commit()

            mycursor.close() 
        except:
            print("error table Puissance ou  déjà complétée")
            

    
    #------------------------------------------------------------
    # Table: Emmissions
    #------------------------------------------------------------
    elif table_name=="Emmissions":
        data=data.set_index([pd.Index([i for i in range(data.shape[0])])])
        try:
            for i in data.index.to_list():
                Co2=data.loc[i,"Co2"]
                CO=data.loc[i,"CO"]
                HC=data.loc[i,"HC"]
                NOx=data.loc[i,"NOx"]
                HC_NOX=data.loc[i,"HC_NOX"]
                Particules=data.loc[i,"Particules"]
                
                command=f'''insert into Emmissions(ID_Emmissions,Co2, CO, HC, NOx, HC_NOX,Particules ) Values({i},{Co2},{CO},{HC},{NOx},{HC_NOX},{Particules})'''
               
                mycursor.execute(command)
                mydb.commit()
                
            mycursor.close()     
        except:
            print("error table Emmissions ou  déjà complétée")
    
    #------------------------------------------------------------
    # Table: Carrosserie
    #------------------------------------------------------------
    elif table_name=="Carrosserie":
        data=data.set_index([pd.Index([i for i in range(data.shape[0])])])
        try:
            for i in data.index.to_list():
                Carrosserie=data.loc[i,"Carrosserie"]
                command=f'''insert into Carrosserie(ID_Carrosserie,Carrosserie) Values({i},"{Carrosserie}")'''
                mycursor.execute(command)
                mydb.commit()
              
            mycursor.close()     
        except:
            print("error table Carrosserie ou  déjà complétée")
            
    #------------------------------------------------------------
    # Table: Voiture
    #------------------------------------------------------------
    elif table_name=="Voiture":
        data=data.set_index([pd.Index([i for i in range(data.shape[0])])])
        #try:
        for i in data.index.to_list():
            Cnit=data.loc[i,"Cnit"]
            ID_Gamme=data.loc[i,"ID_Gamme"]
            ID_Carrosserie=data.loc[i,"ID_Carrosserie"]
            ID_Energie=data.loc[i,"ID_Energie"]
            ID_Marque=data.loc[i,"ID_Marque"]
            ID_Emmissions=data.loc[i,"ID_Emmissions"]
            ID_Puissance=data.loc[i,"ID_Puissance"]
            ID_Consommation=data.loc[i,"ID_Consommation"]
            
            command=f'''insert into Voiture(Cnit,ID_Gamme, ID_Carrosserie, ID_Energie, ID_Marque, ID_Emmissions,ID_Puissance,ID_Consommation ) Values("{Cnit}",{ID_Gamme},{ID_Carrosserie},{ID_Energie},{ID_Marque},{ID_Emmissions},{ID_Puissance},{ID_Consommation})'''
            
            mycursor.execute(command)
            mydb.commit()

        mycursor.close() 
        #except:
        # all  print("error table Voiture ou  déjà complétée")
```
